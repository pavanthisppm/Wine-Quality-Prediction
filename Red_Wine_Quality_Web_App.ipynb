**Create a file named app.py in the same working directory to write code for streamlit**

```python
%%writefile app.py
#import libraries
import streamlit as st
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

#load the dataset & the model
da = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv" ,sep=';')
da.columns = da.columns.str.replace(' ', '_')
da.drop_duplicates(inplace=True)
da.reset_index(drop=True, inplace=True)
da['qualityN']=pd.cut(da['quality'],bins=[0,5.5,10], labels=[0,1])
X_train, X_test, Y_train, Y_test = train_test_split(da.drop(['qualityN','quality'],axis='columns'),da.qualityN,test_size=0.2)
model = RandomForestClassifier()
model.fit(X_train.values,Y_train.values)

st.title("Red Wine Quality Prediction App")

  
#take user inputs
def user_input_features():
        fixed_acidity = st.number_input('Fixed Acidity' ,step=0.001)
        volatile_acidity = st.number_input('Volatile Acidity',step=0.001 )
        citric_acid = st.number_input('Citric Acid' ,step=0.001)
        residual_suagar=st.number_input('Residual Sugar',step=0.001 )
        chlorides = st.number_input('Chlorides' ,step=0.001)
        free_sulfur_dioxide=st.number_input('Free Sulphur Dioxide',step=0.001 )
        total_sulfur_dioxide=st.number_input('Total Sulfur Dioxide',step=0.001 )
        density=st.number_input('Density',step=0.001 )
        pH=st.number_input('pH',step=0.001 )
        sulphates=st.number_input('Sulphates' ,step=0.001 )
        alcohol=st.number_input('Alcohol' ,step=0.001)
        data = {'fixed_acidity': fixed_acidity,
                'volatile_acidity': volatile_acidity,
                'citric_acid': citric_acid,
                'residual acid':residual_suagar,
                'chlorides': chlorides,
                'free_sulfur_dioxide':free_sulfur_dioxide,
                'total_sulfur_dioxide':total_sulfur_dioxide,
                'density':density,
                'pH':pH,
                'alcohol':alcohol,
                'sulphates':sulphates}
        features = pd.DataFrame(data,index=[0])
        return features
df = user_input_features()


#prediction
prediction = model.predict(df.values)
st.subheader('Prediction')
if(prediction == 1):
  st.write('Congartulations! A good quality wine')
else:
  st.write('A bad quality wine. Try next time')
```
